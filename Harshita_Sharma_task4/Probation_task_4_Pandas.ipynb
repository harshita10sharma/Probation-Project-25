####Pandas Data  Analysis

```python
stud = pd.DataFrame({"Name":["Akshat", "Meera", "Shivani","Avantika","Piyush","Rohit","Bhawana","Nidhi","Subh","Mansi"], "Maths":[random.randint(50,100) for _ in range(10)],"Science":[random.randint(50,100) for _ in range(10)],"English":[random.randint(50,100) for _ in range(10)]})
stud
```

```python
stud[["Maths","Science","English"]].mean()
```

```python
stud["Total"] = stud["Maths"]+stud["Science"]+stud["English"]
stud
```

```python
stud["Total"].max()
```

```python
high= stud[stud["Total"]==stud["Total"].max()]
high
```

```python
stud.sort_values("Total",ascending=False)
```

```python
stud['Result']="Fail"
stud.loc[stud['Total']>=150,"Result"] = "Pass"
stud
```
